Importing the Dependencies

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from xgboost import XGBRegressor
from sklearn import metrics
```

Data Collection & Processing

```python
# loading the data from csv file to a Pandas DataFrame
calories = pd.read_csv('/content/calories.csv')
```

```python
# print the first 5 rows of the dataframe
calories.head()
```

```python
exercise_data = pd.read_csv('/content/exercise.csv')
```

```python
exercise_data.head()
```

Combining the two Dataframes

```python
dataset = pd.concat([exercise_data, calories['Calories']], axis=1)
```

```python
dataset.head()
```

```python
# checking the number of rows and columns
dataset.shape
```

```python
# getting some informations about the data
dataset.info()
```

```python
# checking for missing values
dataset.isnull().sum()
```

```python
# get some statistical measures about the data
dataset.describe()
```

Data Visualization

```python
sns.set()
```

```python
# plotting the gender column in count plot
sns.countplot(dataset['Gender'])
```

```python
# finding the distribution of "Age" column
sns.displot(dataset['Age'])
```

```python
# finding the distribution of "Height" column
sns.displot(dataset['Height'])
```

```python
# finding the distribution of "Weight" column
sns.displot(dataset['Weight'])
```

Finding the Correlation in the dataset
- Positive Correlation
- Negative Correlation

```python
c = dataset.corr()
```

```python
# constructing a heatmap to understand the correlation
plt.figure(figsize = (10, 10))
sns.heatmap(c, cbar = True, square = True, fmt = '.1f', annot = True, annot_kws = {'size' : 8}, cmap = 'Blues')
```

Converting the text data to numerical values

```python
dataset.replace({"Gender" : {'male' : 0, 'female' : 1}}, inplace = True)
```

```python
dataset.head()
```

```python
X = dataset.drop(columns = ['User_ID','Calories'], axis = 1)
Y = dataset['Calories']
print(X)
print(Y)
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, random_state=2)

print(X.shape, X_train.shape, X_test.shape)
```

Model Training: XGBoost Regressor

```python
# loading the model
m = XGBRegressor()
```

```python
# training the model with X_train
m.fit(X_train, Y_train)
```

Evaluation - Prediction on Test Data

```python
test_data_predict = m.predict(X_test)

print(test_data_predict)
```

Mean Absolute Error

```python
mae = metrics.mean_absolute_error(Y_test, test_data_predict)

print("Mean Absolute Error = ", mae)
```
